```python
'''
Task 1
'''
def valid_length(password):
    return len(password) >= 8


#unit tests for task1
assert(valid_length('xyz12345') == True)
assert(valid_length('secret123456') == True)
assert(valid_length('') == False)
assert(valid_length('abc123') ==  False)

print('valid_length tests passed')
```

```python
'''
Task 2
'''
def too_obvious(password):
    return password == '12345678' or password.lower() == 'password'

#unit tests for task2
assert(too_obvious('12345678') == True)
assert(too_obvious('password') == True)
assert(too_obvious('PassWord') == True)
assert(too_obvious('PASSWORD') == True)
assert(too_obvious('supersecret') == False)
assert(too_obvious('mypassword') == False)

print('too_obvious tests passed')
```

```python
'''
Task 3
'''
def contains_upper(password):
    for character in password:
        if (character.isupper()):
            return True
    return False


def contains_lower(password):
    for character in password:
        if (character.islower()):
            return True
    return False

#unit tests for contains_upper
assert(contains_upper('passWorD') == True)
assert(contains_upper('1234Abc') == True)
assert(contains_upper('noupper') == False)
assert(contains_upper('567890') == False)

print('contains_upper tests passed')

#unit tests for contains_lower
assert(contains_lower('passWorD') == True)
assert(contains_lower('1234Abc') == True)
assert(contains_lower('NOLOWER') == False)
assert(contains_lower('567890') == False)

print('contains_lower tests passed')
```

```python
'''
Task 4
'''
def contains_digit(password):
    for character in password:
        if (character.isdigit()):
            return True
    return False

def contains_non_digit(password):
    #ok if students use a loop instead of just doing this:
    return not password.isdigit()



#unit tests for contains_digit
assert(contains_digit('pass3word') == True)
assert(contains_digit('abc12345') == True)
assert(contains_digit('0') == True)
assert(contains_digit('secret9') == True)

assert(contains_digit('nodigit') == False)
assert(contains_digit('x') == False)
assert(contains_digit('?!test*') == False)

print('contains_digit tests passed')

#unit tests for contains_non_digit
assert (contains_non_digit('pass3word') == True)
assert(contains_non_digit('abc_def') == True)
assert(contains_non_digit('abc12345') == True)
assert(contains_non_digit('!') == True)
assert(contains_non_digit('0') == False)
assert(contains_non_digit('123456') == False)

print('contains_non_digit tests passed')
```

```python
'''
Task 5
'''
def contains_special(password):
    specialChars = "!#?"
    for c in password:
        if (c in specialChars):
            return True
    return False

#unit tests for contains_special
assert(contains_special('hello!') == True)
assert(contains_special('huh?') == True)
assert(contains_special('#hash') == True)
assert(contains_special('x+y') == False)
assert(contains_special('abc$def') == False)
assert(contains_special('nothing special here.') == False)
assert(contains_special('123456') == False)

print('contains_special tests passed')
```

```python
'''
Task6
''' 
def strength_count(password):
    count = 0 
    if (valid_length(password)):
        count += 1
    if (too_obvious(password) == False):
        count += 1
    if (contains_upper(password)):   
        count += 1
    if (contains_lower(password)):   
        count += 1
    if (contains_digit(password)):   
        count += 1
    if (contains_non_digit(password)):   
        count += 1
    if (contains_special(password)):   
        count += 1
    return count
    
#unit tests for strength_count
assert(strength_count('Secret123!') == 7) #length, not obvious, digit, non digit, upper, lower, special
assert(strength_count('SuperSecret123') == 6) #length, not obvious, digit, non digit, upper, lower
assert(strength_count('SuperSecret') == 5)#length, not obvious, non digit, upper, lower
assert(strength_count('supersecret') == 4)#length, not obvious, non digit, lower
assert(strength_count('secret') == 3) #not obvious, non digit,  lower
assert(strength_count('password') == 3)  #lower, length, non-digit
assert(strength_count('12345678') == 2)  #length, digit
assert(strength_count('1234') == 2)  #not obvious, digit

print('strength_count tests passed')
```

```python
'''
Task 7
'''
def strength_rating(password):
    count = strength_count(password)
    if count == 7:  
        return 'very strong'
    elif count >= 5:
        return 'strong'
    elif count >=3: 
        return 'weak'
    else:
        return 'very weak'
    
#unit tests for strength_rating
assert(strength_rating('Secret123!') == 'very strong')
assert(strength_rating('SuperSecret123') == 'strong')
assert(strength_rating('SuperSecret') == 'strong') 
assert(strength_rating('supersecret') == 'weak') 
assert(strength_rating('secret') == 'weak')  
assert(strength_rating('password') == 'weak')   
assert(strength_rating('12345678') == 'very weak')  
assert(strength_rating('1234') == 'very weak')  

print('strength_rating tests passed')
```
