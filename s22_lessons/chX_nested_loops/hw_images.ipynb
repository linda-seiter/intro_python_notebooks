```python
'''
Task1
Create a new 200x200 image with default color
Use a nested loop to set rgb of each pixel.
Set the green to a random value between 0 and 255 in the left half of the image (red and blue 0),
and the blue to a random value in the right half of the image (red and green 0).
'''


image = Image.new(mode = "RGB", size = (200, 200))

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(rows):
    for c in range(cols):
        random_num = random.randint(0,255)
        if (c<=cols//2):
            pixels_2D[r][c] = [0,random_num,0]
        else:
            pixels_2D[r][c] = [0,0,random_num]
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python
'''
Task2 - 
Create 255x255 image.
First row pixel green should be 255.  First column pixel red should be 255.  Amount of blue is always 0.
Green should decrease by 1 each row.
Red decrease by 1 each column.  Note that red should reset to 255 for each new row. 

'''

image = Image.new(mode = "RGB", size = (255, 255))

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])
green = 255
red = 255
blue = 0
for r in range(rows):
    for c in range(cols):
        pixels_2D[r][c] = [red, green, blue]
        red -= 1
    green -= 1
    red = 255
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python
'''
Task3 - 
Step 1, use a nested loop to produce a mirror image of the bird.
be careful about indexing and off by 1 errors 
(if width is 180, indices are between 0 and 179)
'''

image = Image.open('bird.jpg')
print(image)
display(image)

#remove red from pixels in top half
pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(rows//2):
    for c in range(cols//2):
        pixel = pixels_2D[r][c]
        #mirror 4 quadrants
        pixels_2D[r][cols-c-1] = pixel
        pixels_2D[rows-r-1][c] = pixel
        pixels_2D[rows-r-1][cols-c-1] = pixel
        
image = Image.fromarray(pixels_2D)
display(image)
```
