```python
#you can get all of the pixels in the image as a 1D list
image = Image.open('cat.jpg')
pixels_1D = list(image.getdata())
print(type(pixels_1D))
print(len(pixels_1D))

first_pixel = pixels_1D[0]
print(first_pixel)  #pixel color is a tuple (r,g,b), range is 0 to 255
print(type(first_pixel))

#use indexing to get the r,g,b values from the pixel
print(first_pixel[0])  #r
print(first_pixel[1])  #g
print(first_pixel[2])  #b
```

```python
#reload the original image
image = Image.open('cat.jpg')
pixels_1D = list(image.getdata())

#change color of first 1000 pixels to red.
for p in range(1000):
    pixels_1D[p] = (255,0,0)  #rgb

image.putdata(pixels_1D)  #update the image to use the modified pixels
display(image)
```

```python
#remove all red from each pixel in the top half of the image

#reload the original image
image = Image.open('cat.jpg')
display(image)

pixels_1D = list(image.getdata())
for p in range(len(pixels_1D)//2):
    next_pixel = pixels_1D[p]   #(r,g,b)
    red=0
    green=next_pixel[1]
    blue=next_pixel[2]
    pixels_1D[p] = (red,green,blue)
    
image.putdata(pixels_1D)  #update the image to use the modified pixels
display(image)
```

```python
#set every nth pixel to black
#reload the original image

image = Image.open('cat.jpg')
display(image)

pixels_1D = list(image.getdata())
n=int(input('enter n:'))

for p in range(0,len(pixels_1D),n):
    pixels_1D[p] = (0,0,0)  #black
    
image.putdata(pixels_1D)  
display(image)
```

```python
#Using numpy to manipulate image as 2D (rows, columns)

image = Image.open('cat.jpg')
display(image)

pixels_2D = np.array(image)  #convert to 2D list
print(len(pixels_2D))  #height, rows
print(len(pixels_2D[0])) #width, columns
print(pixels_2D[0][0])  #first pixel, notice it is a list not tuple
pixels_2D[10][100] = [0, 50, 100]  #change color of pixel[r][c] to black
image = Image.fromarray(pixels_2D)
display(image)
```

```python
image = Image.new(mode = "RGB", size = (400, 400))

#increase amount of red for each row

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(rows):
    for c in range(cols):
        if (r<=255):
            pixels_2D[r][c] = [r,0,0]
        else:
            pixels_2D[r][c] = [255,0,0]
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python

#getpixel function returns a pixel at a given (x,y) tuple 
#putpixel sets pixel value at (x,y) tuple
image = Image.open('cat.jpg')

xy=(100,10)  #note this is a tuple, not a list
pixel = image.getpixel(xy)   #NOTE: pass a tuple (x,y) as parameter
print(pixel)
pixel=(0,0,0) #black
image.putpixel(xy, pixel)  #parameters are coordinate tuple and pixel tuple
display(image)
```

```python
image = Image.open('cat.jpg')

#copy left half to right half

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])
half_width = cols//2
for r in range(rows):
    for c in range(cols//2):
        left_pixel = pixels_2D[r][c]
        pixels_2D[r][c+half_width] = left_pixel
        
image = Image.fromarray(pixels_2D)
display(image)
```
