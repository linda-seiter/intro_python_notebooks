## 2D image manipulation

```python
#Using numpy to manipulate image as 2D (rows, columns)

image = Image.open('cat.jpg')
display(image)

pixels_2D = np.array(image)  #convert to 2D list
print(len(pixels_2D))  #height, rows
print(len(pixels_2D[0])) #width, columns
print(pixels_2D[0][0])  #first pixel, notice it is a list not tuple
pixels_2D[10][100] = [0, 50, 100]  #change color of pixel[r][c] to black
image = Image.fromarray(pixels_2D)
display(image)
```

```python
#set all pixels in first 20 rows to red
image = Image.open('cat.jpg')
display(image)

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(20):
    for c in range(cols):
        pixels_2D[r][c] = [255,0,0]  #reassign pixel rgb to max red of 255
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python
image = Image.open('cat.jpg')
display(image)

#remove red from pixels in first 50 rows
pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(50):
    for c in range(cols):
        pixel = pixels_2D[r][c]
        red=0
        green = pixel[1]
        blue = pixel[2]
        pixels_2D[r][c] = [red,green,blue]
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python
image = Image.open('cat.jpg')
display(image)

#add green stripe every 10th row
pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(0,rows,10):
    for c in range(cols):
        pixels_2D[r][c] = [0,255,0]
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python
image = Image.new(mode = "RGB", size = (400, 400))

#increase amount of red for each row

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])

for r in range(rows):
    for c in range(cols):
        if (r<=255):
            pixels_2D[r][c] = [r,0,0]
        else:
            pixels_2D[r][c] = [255,0,0]
        
image = Image.fromarray(pixels_2D)
display(image)
```

```python

#getpixel function returns a pixel at a given (x,y) tuple 
#putpixel sets pixel value at (x,y) tuple
image = Image.open('cat.jpg')

xy=(100,10)  #note this is a tuple, not a list
pixel = image.getpixel(xy)   #NOTE: pass a tuple (x,y) as parameter
print(pixel)
pixel=(0,0,0) #black
image.putpixel(xy, pixel)  #parameters are coordinate tuple and pixel tuple
display(image)
```

```python
image = Image.open('cat.jpg')

#copy left half to right half

pixels_2D = np.array(image)  #convert to 2D list
rows = len(pixels_2D)
cols = len(pixels_2D[0])
half_width = cols//2
for r in range(rows):
    for c in range(cols//2):
        left_pixel = pixels_2D[r][c]
        pixels_2D[r][c+half_width] = left_pixel
        
image = Image.fromarray(pixels_2D)
display(image)
```
