```python
# interact generates UI controls and calls biggest method when either control changed

def biggest(a, b):
    if a>b:
      return a
    else:
      return b

interact(biggest, 
         a=widgets.IntSlider(min=0,max=50,step=5,value=10), 
         b=widgets.IntSlider())
```

```python
# interact_manual generates a button to trigger interaction

def biggest(a, b):
    if a>b:
      return a
    else:
      return b

interact_manual(biggest, 
         a=widgets.IntSlider(min=0,max=50,step=5,value=10), 
         b=widgets.IntSlider())
```

```python
#Define interaction using @interact decorator

def biggest(a, b):
    if a>b:
      return a
    else:
      return b

@interact
def setupUI(num1 = (1,100), 
            num2 = (1,100)):
  print(biggest(num1, num2))
```

```python
#widgets default to continuous update

@interact
def demo(num1=(0, 10, 1), 
         num2=(0, 100, 5), 
         isRaining=True, 
         fruit=["apple","banana","orange"]):
    print("sum ",num1+num2)
    print("is is raining? " , isRaining)
    print("favorite fruit ", fruit)
```

```python
#interact_manual button delays reaction

@interact_manual
def demo2(num1=(0, 10, 1), 
          num2=(0, 10, 5), 
          isRaining=True, 
          fruit=["apple","banana","orange"]):
    print("sum ",num1+num2)
    print("is is raining? " , isRaining)
    print("favorite fruit ", fruit)
```

```python
#textbox and slider

def repeat_word(word, times):
    return (word+" ")*times

interact(repeat_word, word="Hello",  times=(0,100,5))
```
